```python
# converting categorical columns into numerical using label encoding
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
# train data
for feature in sdtrain1.columns:
  if sdtrain1[feature].dtype == 'object':
    sdtrain1[feature] = encoder.fit_transform(sdtrain1[feature])
# test data
for feature in sdtest1.columns:
  if sdtest1[feature].dtype == 'object':
    sdtest1[feature] = encoder.fit_transform(sdtest1[feature])
```

```python
sdtrain1.head()
```

```python
# separting target variable from data(train and test)
X_train = sdtrain1.iloc[:,:-1]
y_train = sdtrain1.iloc[:,-1]
X_test = sdtest1.iloc[:,:-1]
y_test = sdtest1.iloc[:,-1]
```

#Model training
#Gaussian Naive Bayes

```python
# training model
train_model = GaussianNB()
train_model.fit(X_train,y_train)
```

```python
print("train score",train_model.score(X_train,y_train))
print("test score",train_model.score(X_test,y_test))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,train_model.predict(X_test)))
```

```python
#Multinominal Naive Bayes
mb_model = MultinomialNB()
mb_model.fit(X_train,y_train)
```

```python
y_mb_train = mb_model.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
# training data
accuracy_score(y_train,y_mb_train)
```

```python
confusion_matrix(y_train,y_mb_train)
```

```python
print(classification_report(y_test,mb_model.predict(X_test)))
```

```python
from sklearn.naive_bayes import CategoricalNB

rf_clf = CategoricalNB(alpha=1.0)
rf_clf.fit(X_train, y_train)
```

```python
print(classification_report(y_test,rf_clf.predict(X_test)))
```
